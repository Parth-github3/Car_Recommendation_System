# Cosine Similarity

```python
car_df['Year']=car_df['Year'].apply(str)
car_df['km']=car_df['km'].apply(str)
car_df['Price']=car_df['Price'].apply(str)
car_df['tags'] = car_df[['Make', 'Model', 'Year', 'km',	'Price','regCity',	'Assembly',	'Transmission',	'sellerloc', 'carFeatures']].agg(' '.join, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
car_df = car_df.apply(lambda col: col.str.lower() if col.dtype == 'object' else col)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(car_df['tags'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

Make=car_df['Make']
tag= car_df['tags']
def recommend_cars(Make_index):
      similar_cars = list(enumerate(cosine_sim[Make_index]))
      similar_cars = sorted(similar_cars, key=lambda x: x[1], reverse=True)[1:10] 
      car_indices = [i[0] for i in similar_cars]
      return car_df['Make'].iloc[car_indices]
```
